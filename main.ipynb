```python
movies.dropna(subset=['title', 'genres', 'cast', 'crew'],inplace=True)
```

```python
movies['overview'] = movies['overview'].fillna('')
movies['keywords'] = movies['keywords'].fillna('')
movies['tagline'] = movies['tagline'].fillna('')
```

```python
movies['genres'].head(2)
```

```python
import ast

# Hàm lấy chỉ tên của các thể loại / ....
def convert(text):
    l =[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies['cast'].head()
```

```python
def convert_cast(text):
    l =[]
    count = 0
    for i in ast.literal_eval(text):
        if count < 3:
            l.append(i['name'])
            count +=1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies['cast'].head()
```

```python
movies['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords']
```

```python
def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
            
    return l
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies['overview']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Xóa khoảng trắng
def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(" ",""))
    return l 
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
def split_words(text):
    l = []
    for i in text:
        l.append(i.replace(' ',','))
    return l
```

```python
## GÁN TAG CHO MOVIES
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
df = movies[['id','title','tags']]
df.head()
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
        
    return " ".join(l)
```

```python
df['tags'] = df['tags'].apply(stems)
```

```python
df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words= 'english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similary = cosine_similarity(vector)
```

```python
similary
```

```python
similary.shape
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similary[index])), reverse=True, key = lambda x:x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```
