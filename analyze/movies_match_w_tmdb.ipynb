```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
movies = pd.read_csv('../data/movies_metadata.csv')
links = pd.read_csv('../data/links_small.csv')
```

```python
print("movies:", movies.shape)
print("links :", links.shape)
print("unique tmdbId:", links['tmdbId'].nunique())
```

```python
# ---- LINKS ----
links['tmdbId'] = pd.to_numeric(links['tmdbId'], errors='coerce')
links = links.dropna(subset=['tmdbId'])
links['tmdbId'] = links['tmdbId'].astype(int)

# check lại
links['tmdbId'].dtype
```

```python
movies['id'] = pd.to_numeric(movies['id'], errors='coerce')
movies = movies.dropna(subset=['id'])
movies['id'] = movies['id'].astype(int)

movies = movies[movies['id'].isin(links['tmdbId'])]
```

```python
movies.shape
```

```python
# Tạo thư mục data nếu chưa tồn tại
data_dir = os.path.join("..", "data")
os.makedirs(data_dir, exist_ok=True)

# Lưu DataFrame movies vào file CSV

# Đường dẫn file CSV
movies_match_tmdb = os.path.join(data_dir, "movies_match_tmdb.csv")
movies.to_csv(movies_match_tmdb, index=False, encoding='utf-8-sig')
```
